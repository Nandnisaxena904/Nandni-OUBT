Create a YAML metadata manifest using Python (very common)
YAML is more readable and often preferred for governance.

```python
import yaml
from datetime import datetime, timezone

metadata = {
    "bucket_name": "nandni-oubt",
    "purpose": "Stores raw and curated data for practicing Data engineering with AWS during OUBT internship",
    "owner": {
        "team": "OUBT Data Engineering",
        "email": "nandnisaxena904@gmail.com"
    },
    "data_domain": "Training",
    "classification": "Confidential",
    "technical_metadata": {
        "storage_format": ["csv", "parquet"],
        "region": "us-west-2",
        "versioning_enabled": True
    },
    "access_governance": {
        "read_access": ["AnalyticsRole", "FinanceRole"],
        "write_access": ["DataEngineeringRole"],
        "pii_present": True
    },
    "created_date": datetime.now(timezone.utc).isoformat()
}




with open("metadata_manifest.yaml", "w") as f:
    yaml.dump(metadata, f, sort_keys=False)

print("metadata_manifest.yaml created")
```

STORE THIS IN S3

```python
import boto3

s3 = boto3.client("s3")

s3.upload_file(
    "metadata_manifest.yaml",
    "nandni-oubt",
    "_metadata/manifest.yaml"
)

print("Manifest uploaded to S3")
```
